```python
import torch
from torch.utils.data import DataLoader

def _dataloader_dist(train_set, train_sampler, batch_size, collator):
        torch.multiprocessing.set_start_method('spawn', force=True)
        train_dataloader = DataLoader(
            train_set,
            sampler=train_sampler,
            pin_memory=False,
            num_workers=2,
            persistent_workers=True, 
            drop_last=True,
            batch_size=batch_size,
            collate_fn=collator,
        )
        return train_dataloader
```

```python
import torch


new_tuples = []
```

```python
for original_tuple in train_set:
    values_list = []
    for value in original_tuple[0].values():
        values_list.append(value)
    values_tensor = torch.tensor(values_list)
    indices_tensor = torch.tensor(original_tuple[1])
    new_tuple = (values_tensor, indices_tensor)
    new_tuples.append(new_tuple)
```

```python
new_tuples
```

```python
merged_samples = []
for sample in new_tuples:
    merged_tensor = torch.cat(sample, dim=0)
    merged_samples.append(merged_tensor)

merged_tuple_list = [torch.chunk(sample, chunks=2, dim=0) for sample in merged_samples]

print(merged_tuple_list)
```

```python
tensor_list = []
for sample in new_tuples:
    new = torch.cat((sample[0], sample[1]), dim=0)
    
    tensor_list.append(new)
    
```

```python
from torch.nn.utils.rnn import pad_sequence
```

```python
# 对齐样本到最大序列长度，并使用0进行填充
padded_sequence = pad_sequence(tensor_list, batch_first=True, padding_value=0)

print(padded_sequence)
```
